```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston

import seaborn as sns
```

```python
data = pd.DataFrame(load_boston().data)
data
```

```python
target = pd.DataFrame(load_boston().target)
target.head()
```

```python
df = pd.concat((data, target), axis = 1)
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

df = StandardScaler().fit_transform(df)
df = pd.DataFrame(df)
df.head()
```

```python
sns.boxplot(x = 'variable', y = 'value', data = df.melt())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.33, random_state=0)
```

```python
display(data.shape)
display(target.shape)
display(X_train.shape)
display(X_test.shape)
display(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

print(f'Train Acc: {lr_model.score(X_train, y_train)}')
print(f'Test ACc: {lr_model.score(X_test, y_test)}')
```

```python
from sklearn.linear_model import Lasso

param = []
train = []
test = []
for i in [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5]:

    lasso_model = Lasso(i)
    lasso_model.fit(X_train, y_train)
    
    print(f'L2 Param: {i}')
    print(f'Train Acc: {lasso_model.score(X_train, y_train)}')
    print(f'Test ACc: {lasso_model.score(X_test, y_test)} \n')
    
    param.append(i)
    train.append(lasso_model.score(X_train, y_train))
    test.append(lasso_model.score(X_test, y_test))
    
plt.plot(param, train, 'b')
plt.plot(param, test, 'r')
plt.show()

print(lasso_model.coef_)
```

```python
from sklearn.linear_model import Ridge

param = []
train = []
test = []
for i in [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5]:

    ridge_model = Ridge(i)
    ridge_model.fit(X_train, y_train)
    
    print(f'L2 Param: {i}')
    print(f'Train Acc: {ridge_model.score(X_train, y_train)}')
    print(f'Test ACc: {ridge_model.score(X_test, y_test)} \n')
    
    param.append(i)
    train.append(ridge_model.score(X_train, y_train))
    test.append(ridge_model.score(X_test, y_test))
    
plt.plot(param, train, 'b')
plt.plot(param, test, 'r')
plt.show()

print(ridge_model.coef_)
```
